Visualizations

Salary Distributions

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
# Preparing data for visualizations
# For salary distribution by job title, let's select the top 10 job titles based on frequency
top_job_titles = data['job_title'].value_counts().head(10).index

# Data for salary distribution by experience level
experience_order = ['Entry-level', 'Mid-level', 'Senior', 'Executive']
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data[data['job_title'].isin(top_job_titles)], x='salary_in_usd', y='job_title')
plt.title('Salary Distribution by Job Title')
plt.xlabel('Salary in USD')
plt.ylabel('Job Title')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='salary_in_usd', y='experience_level', order=experience_order)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Salary in USD')
plt.ylabel('Experience Level')
plt.tight_layout()
plt.show()
```

```python
# Defining top_countries and top_categories
top_countries = data['employee_residence'].value_counts().nlargest(10).index
top_categories = data['job_category'].value_counts().nlargest(10).index

# Average salary by job title for the top 10 most frequent job titles
top_job_titles = data['job_title'].value_counts().nlargest(10).index
top_job_titles_avg = data[data['job_title'].isin(top_job_titles)].groupby('job_title')['salary_in_usd'].mean()

plt.figure(figsize=(10, 8))
sns.boxplot(data=data[data['employee_residence'].isin(top_countries)], x='salary_in_usd', y='employee_residence')
plt.title('Geographic Distribution of Salaries (Top 10 Countries)')
plt.xlabel('Salary in USD')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='salary_in_usd', y='employment_type')
plt.title('Comparison of Salary by Employment Type')
plt.xlabel('Salary in USD')
plt.ylabel('Employment Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='salary_in_usd', y='company_size', order=['S', 'M', 'L'])
plt.title('Salary Distribution by Company Size')
plt.xlabel('Salary in USD')
plt.ylabel('Company Size')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data[data['job_category'].isin(top_categories)], x='salary_in_usd', y='job_category')
plt.title('Salary Distribution by Job Category')
plt.xlabel('Salary in USD')
plt.ylabel('Job Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data['salary_in_usd'], bins=30, kde=True)
plt.title('Histogram of Salary Distributions')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_job_titles_avg.plot(kind='bar')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
experience_job_count = data.groupby(['experience_level', 'job_title']).size().reset_index(name='counts')
experience_job_count_pivot = experience_job_count.pivot_table(index="job_title", columns="experience_level", values="counts", fill_value=0)
top_experience_jobs = experience_job_count_pivot.sum(axis=1).sort_values(ascending=False).head(10).index
experience_job_count_pivot.loc[top_experience_jobs].plot(kind='bar', stacked=True)
plt.title('Count of Job Titles by Experience Level')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.legend(title='Experience Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_salary_by_year = data.groupby('work_year')['salary_in_usd'].mean()
plt.figure(figsize=(12, 8))
average_salary_by_year.plot(kind='bar')
plt.title('Average Salary by Year')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
top_countries_salary = data[data['employee_residence'].isin(data['employee_residence'].value_counts().head(5).index)]

# Prepare data for the dual-axis graph on job categories
job_category_stats = data.groupby('job_category').agg({'salary_in_usd': ['mean', 'count']})
job_category_stats.columns = ['Average Salary', 'Job Count']
job_category_stats = job_category_stats.sort_values(by='Average Salary', ascending=False)

# Calculate Yearly Salary Growth Rate
yearly_average_salary = data.groupby('work_year')['salary_in_usd'].mean()
yearly_growth_rate = yearly_average_salary.pct_change() * 100
```

```python
plt.figure(figsize=(10, 8))
sns.kdeplot(data=top_countries_salary, x="salary_in_usd", hue="employee_residence", fill=True)
plt.title('Salary Distribution by Geographic Region')
plt.xlabel('Salary in USD')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax1 = job_category_stats['Average Salary'].plot(kind='bar', color='skyblue')
ax2 = ax1.twinx()
job_category_stats['Job Count'].plot(kind='line', marker='o', color='green', ax=ax2)
plt.title('Average Salary and Job Count by Job Category')
ax1.set_xlabel('Job Category')
ax1.set_ylabel('Average Salary in USD', color='skyblue')
ax2.set_ylabel('Job Count', color='green')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(data=data, x='experience_level', y='salary_in_usd')
plt.title('Experience Level Salary Distributions')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
yearly_growth_rate.plot(kind='line', marker='o')
plt.title('Yearly Salary Growth Rate')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Plot average salary
ax1 = job_category_stats['Average Salary'].plot(kind='bar', color='skyblue', position=0, width=0.4)
ax2 = ax1.twinx()
job_category_stats['Job Count'].plot(kind='bar', color='orange', ax=ax2, position=1, width=0.4)

# Set titles and labels
plt.title('Average Salary and Job Count by Job Category')
ax1.set_xlabel('Job Category')
ax1.set_ylabel('Average Salary in USD')
ax2.set_ylabel('Job Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
# Define the top categories dynamically
top_categories = data['job_category'].value_counts().nlargest(10).index.tolist()

# Enhanced Geographic Distribution of Salaries with Dropdown
fig1 = px.box(data[data['employee_residence'].isin(top_countries)], 
              x='salary_in_usd', y='employee_residence',
              color='employee_residence',  # Color by residence for better distinction
              title='Geographic Distribution of Salaries (Top 10 Countries)', 
              labels={'salary_in_usd': 'Salary in USD', 'employee_residence': 'Employee Residence'},
              color_discrete_sequence=px.colors.qualitative.Set2)  # Beautiful color palette



# Show plots
fig1.show()
```

```python
# Adding dropdown for job categories in Salary Distribution plot
fig2 = go.Figure()
for category in top_categories:
    fig2.add_trace(go.Box(x=data[data['job_category'] == category]['salary_in_usd'],
                          name=category,
                          marker_color=px.colors.qualitative.Pastel1[data['job_category'].tolist().index(category) % len(px.colors.qualitative.Pastel1)]))

fig2.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=[{"visible": [category == selected_category for selected_category in top_categories]}],
                    label=category,
                    method="update"
                ) for category in top_categories
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.11,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
    ],
    title="Salary Distribution by Job Category",
    xaxis_title="Salary in USD",
    yaxis_title="Job Category"
)
fig2.show()
```

```python
import plotly.express as px

# Interactive Average Salary by Job Title with Color Gradient
fig3 = px.bar(x=top_job_titles_avg.index, y=top_job_titles_avg.values,
              labels={'x': 'Job Title', 'y': 'Average Salary in USD'},
              title='Average Salary by Job Title',
              color=top_job_titles_avg.values,  # Add color based on salary values
              color_continuous_scale=px.colors.sequential.Viridis)  # Use a continuous color scale
fig3.show()
```
